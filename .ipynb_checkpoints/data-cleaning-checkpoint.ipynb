```python
# Importing the dependencies

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read in the raw data file to a dataframe
file = 'Resources/owid-covid-data.csv'
df = pd.read_csv(file)
df
```

```python
#list(df.columns)
```

```python
# Removed unknown values from Operator and Operator ID
df = df.loc[(df['continent'] != 'NaN')]
df = df.loc[(df['iso_code'] != 'NaN')]
```

```python
# Clean up the dataframe
# Remove all NaN values, replace with 0 for relevant columns
df['total_cases'].fillna(0,inplace = True)
df['total_deaths'].fillna(0,inplace = True)
df
```

```python
# Move cleaned df into csv
df.to_csv("Resources/Cleaned_data.csv",index=False,header=True)
```
